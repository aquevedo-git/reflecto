```python
# Cell 1 — Setup
from reflecto.chatbot.flow import run_reflecto_flow
def fake_answer_callback(prompt):
    return 'Short answer.'
print("Phase 2 setup OK")
```

```python
# Cell 2 — Normal energy flow
user_state = {
    'user_id': 'test_user',
    'energy': 7
}
prompts = run_reflecto_flow(user_state, fake_answer_callback)
for p in prompts:
    print(p)
assert len(prompts) > 2, f'Expected >2 questions, got {len(prompts)}'
print(f'Questions asked: {len(prompts)}')
```

```python
# Cell 3 — Low energy flow
user_state = {
    'user_id': 'test_user',
    'energy': 2
}
prompts = run_reflecto_flow(user_state, fake_answer_callback)
for p in prompts:
    print(p)
assert len(prompts) >= 2, f'Expected at least 2 prompts, got {len(prompts)}'
assert any('gentle' in p.lower() or 'rest' in p.lower() or 'take care' in p.lower() for p in [prompts[-1]]), 'Gentle closing message missing in last prompt'
print(f'Questions asked: {len(prompts) - 1} (excluding closure)')
```
